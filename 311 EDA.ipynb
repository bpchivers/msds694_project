# Quick Column Value exploration

## Opened and Closed

```python
print("Number of unclosed cases: " + str(csv['Closed'].isnull().sum()))
print("Percent of cases that are unclosed: " + str(csv['Closed'].isnull().sum()/len(csv)))
```

```python
csv['Opened'].min()
```

## Status

```python
csv["Status"].unique()
```

## Responsible Agency, Catagory and Request Type

```python
csv[["CaseID","Responsible Agency"]].groupby(["Responsible Agency"]).agg(['count']).sort_values(by=("CaseID","count"),ascending=False)
```

```python
csv[["CaseID","Category"]].groupby(["Category"]).agg(['count']).sort_values(by=("CaseID","count"),ascending=False)
```

```python
csv[["CaseID","Request Type"]].groupby(["Request Type"]).agg(['count']).sort_values(by=("CaseID","count"),ascending=False)
```

## Location based columns

```python
csv[["CaseID","Supervisor District"]].groupby(["Supervisor District"]).agg(['count']).sort_values(by=("CaseID","count"),ascending=False)
```
